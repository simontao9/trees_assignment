```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Question 3.1
data = pd.read_csv("./data/corporate_ratings.csv")
print(data.shape)
data.head()
```

Question 3.1

There are 2029 observations of 31 variables.

```python
# Question 3.2
data["Rating"].unique()
```

```python
def collapse_ratings(rating):
    if pd.isnull(rating):
        return np.nan
    
    if rating.startswith("A"):
        return "A"
    elif rating.startswith("B"):
        return "B"
    elif rating.startswith("C"):
        return "C"
    else:
        return "D"
    
data["Collapsed_Ratings"] = data["Rating"].apply(collapse_ratings)
```

```python
plt.figure()
sns.histplot(data["Collapsed_Ratings"])
plt.xlabel("Rating Class")
plt.show()
```

Question 3.2

Completed in the code chunks above.

```python
# Question 3.3
excluded_vars = ["Rating", "Date", "Name", "Symbol", "Rating Agency Name"]
included_vars = [col for col in data.columns if col not in excluded_vars + ["Collapsed_Ratings"]]

X = pd.get_dummies(data[included_vars], drop_first=True)
y = data["Collapsed_Ratings"]

X.head()
```

Question 3.3

Completed in the code chunk above.

```python
# Question 3.4
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

remove = y.notnull()
X = X[remove]
y= y[remove]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

tree = DecisionTreeClassifier(random_state=0, min_samples_leaf=60)
tree.fit(X_train, y_train)
y_hat = tree.predict(X_test)

confusionMatrix = pd.crosstab(y_test,y_hat) 
accuracy = np.trace(confusionMatrix)/len(y_test) # Faster way to compute accuracy
print(accuracy)
print(confusionMatrix)
```

Question 3.4

I did an 80/20 train/test split of the data. Based on the confusion matrix and model accuracy, the model predicts the different bond ratings with roughly 73% accuracy.

```python
# Question 3.5
included_vars_p5 = [col for col in data.columns if col not in ["Rating", "Date", "Name", "Symbol"]]
X_p5 = pd.get_dummies(data[included_vars_p5], drop_first=True)

X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_p5, y, test_size=0.2, random_state=0)

tree_p5 = DecisionTreeClassifier(random_state=0, min_samples_leaf=70)
tree_p5.fit(X_train_2, y_train_2)
y_hat_2 = tree_p5.predict(X_test_2)

confusionMatrix = pd.crosstab(y_test_2,y_hat_2) 
accuracy = np.trace(confusionMatrix)/len(y_test) # Faster way to compute accuracy
print(accuracy)
print(confusionMatrix)
```
